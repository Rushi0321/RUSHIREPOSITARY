```python
def dummies(x,d1):
    temp=pd.get_dummies(d1[x],drop_first=True)
    d1=pd.concat([d1,temp],axis=1)
    d1.drop([x],axis=1,inplace=True)
    return d1

df2=df
df2=dummies('mainroad',df2)
df2=dummies('guestroom',df2)
df2=dummies('basement',df2)
df2=dummies('hotwaterheating',df2)
df2=dummies('airconditioning',df2)
df2=dummies('prefarea',df2)
df2=dummies('furnishingstatus',df2)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def build_model(X,y):
    X=sm.add_constant(X)
    lm=sm.OLS(y,X).fit()
    print(lm.summary())
    return lm
def checkVIF(X):
    vif=pd.dataframe()
    vif['Features']=X.columns
    vif['VIF']=[variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
    vif['VIF']=round(vif['VIF'],2)
    vif=vif.sort_values(by='VIF', ascending=False)
    return (vif)
```
